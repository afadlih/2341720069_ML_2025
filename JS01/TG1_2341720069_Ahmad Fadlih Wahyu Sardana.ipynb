- Jawaban :

    Di dunia AI sekarang, banyak kasus yang melanggar etika dan hukum, terutama soal bias dan privasi.

    Sistem Rekrutmen yang rasis: Ada perusahaan yang pakai AI buat nyaring calon karyawan, tapi ternyata modelnya bias. Karena model itu dilatih pakai data historis yang mungkin cuma ngerekrut laki-laki atau kelompok tertentu, alhasil sistemnya juga ikutan diskriminatif. Ini jelas melanggar etika keadilan dan bisa digugat karena diskriminasi. (Sumber: UNESCO, Ethics of Artificial Intelligence).

    Penyalahgunaan Data Pribadi: Model AI yang bisa bikin gambar atau teks ada isu di mana model itu dilatih pakai data pribadi orang lain tanpa izin. Jadi, data-data yang ada di internet bisa aja dipakai buat ngelatih AI, dan itu berpotensi melanggar hukum privasi dan hak cipta. (Sumber: Thomson Reuters Legal Solutions, Legal issues with AI: Ethics, risks, and policy).

    Dampak Lingkungan karena AI
  

    Boros Energi dan Air: Melatih model AI canggih itu butuh energi yang besar. Untuk melatih satu model AI bisa ngabisin listrik sebanyak puluhan rumah dalam setahun! Selain itu, pusat data yang nyimpen AI butuh jutaan liter air buat pendinginan. Dampaknya, sumber daya alam jadi terkuras. (Sumber: MIT News, Explained: Generative AI's environmental impact).

    Jejak Karbon yang besar: Karena listrik di banyak tempat masih pakai bahan bakar fosil, konsumsi energi yang tinggi ini otomatis membuat emisi karbon naik. Melatih satu model AI bisa ngeluarin karbon sebanyak mobil yang ada dunia (Sumber: UNEP, AI has an environmental problem. Here's what the world can do about that.).

    Cara Ngatasinnya :

    "Green AI": Kita harus lebih fokus ke AI yang hemat energi. Jadi saat membuat model, jangan mikirin akurasi saja , tapi juga seberapa efisien model itu. Bisa pakai algoritma yang lebih ringan atau optimasi arsitektur.

    Pusat Data Ramah Lingkungan: Perusahaan teknologi harusnya pakai energi terbarukan kayak matahari atau angin buat pusat data mereka. Udah ada kok beberapa perusahaan yang mulai ngatur jadwal kerja AI-nya pas lagi banyak energi surya.

    Riset Hardware Hijau: Kita butuh inovasi di chip dan hardware yang lebih hemat energi dan bisa didaur ulang. Ini penting banget biar sampah elektronik (e-waste) dari komponen AI juga bisa dikurangi.
